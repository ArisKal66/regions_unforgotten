```python
import pandas as pd

df = pd.read_excel('./1952.xls')
df
```

```python
df['MEMB_RATIO'] = df['TOTMEMB']/df['TOTPOP']
ratio = df.groupby(['CNAME'])['MEMB_RATIO'].sum().sort_values(ascending= False)
print(ratio[15:30],'\n')
```

```python
rel_sums = []
for i in range(7,235,2):
    name = df.columns[i]
    value = df.iloc[:3075,i].sum()
    tupple = name,value
    rel_sums.append(tupple)
```

```python
rel_sums = pd.DataFrame(data = rel_sums, columns = ['rel_group','members'])
rel_sums
```

```python
top_group = rel_sums.groupby('rel_group')['members'].sum().sort_values(ascending = False)
print(top_group[0:15],'\n\n',df['TOTPOP'].sum())
```

```python
rel_sums['ratio'] = rel_sums['members']/150635574
rel_sums = rel_sums.sort_values(by=['ratio'], ascending=False)
rel_sums.head(15)
```

```python
orthodox = df.groupby(['CNAME'], as_index= False)[['TOTPOP','ARAPO_M','GRKAD_M','ACROC_M','BEOC_M']].sum()
orthodox['TOTORTH'] = orthodox['ARAPO_M']+orthodox['GRKAD_M']+orthodox['ACROC_M']+orthodox['BEOC_M']
orthodox = orthodox.sort_values(by =['TOTORTH'], ascending = False)
orthodox = orthodox.drop(columns = ['ARAPO_M','GRKAD_M','ACROC_M','BEOC_M'])
orthodox.head(10)
```

```python
rel_dist = []
for j in range(3075):
    name_1 = df['CNAME'][j]
    ch = df['TOTCHUR'][j]
    no_ch = 0
    for l in range(8,236,2):
        if df.iloc[j,l] == 0:
            no_ch +=1
    tupple_1 = name_1,ch,no_ch
    rel_dist.append(tupple_1)
```

```python
rel_dist = pd.DataFrame(data = rel_dist, columns = ['county_name','tot_churches', 'groups_no_ch'])
```

```python
rel_dist = rel_dist.sort_values(by = ['groups_no_ch'])
rel_dist
```
